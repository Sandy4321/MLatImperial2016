## Reading spam collection
Each line is label + TAB + message. Label is either 'spam' or 'ham'. 

Let's look at this file

```python
!head -n 15 ./data/1-sms-spam-train.txt
```

```python
import codecs
with codecs.open('./data/1-sms-spam-train.txt', encoding='utf-8') as f:
    train_labels_textual, train_messages = zip(*[line.split('\t') for line in f.readlines()])
    
with codecs.open('./data/1-sms-spam-test.txt', encoding='utf-8') as f:
    test_messages = f.readlines()
```

## Converting labels to 0 and 1

Just for convenience. 1 is spam, 0 is ham

```python
import numpy
print numpy.array(train_labels_textual)
train_labels = (numpy.array(train_labels_textual) == 'spam') * 1
print train_labels
```

## Computing minimal stats

```python
print 'spam messages', train_labels.sum()
print 'ham messages ', (1 - train_labels).sum()
print 'spam part    ', train_labels.mean()
```

## Defining simple evaluation procedure

```python
def evaluate(message):
    message = message.lower()
    positive_patterns = ['me ', 'he ', 'i ']
    negative_patterns = ['call', 'free', 'won', 'sms',]
    score = 0
    for pattern in positive_patterns:
        score -= (pattern in message)
    for pattern in negative_patterns:
        score += (pattern in message)
    return score
```

```python
train_scores = map(evaluate, train_messages)
```

## Testing the quality on available data

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(train_labels, train_scores)
```

## Preparing the basic solution

```python
predictions = map(evaluate, test_messages)
```

```python
import pandas
result = pandas.DataFrame({'Id': numpy.arange(len(test_messages)), 'Label': predictions})
```
